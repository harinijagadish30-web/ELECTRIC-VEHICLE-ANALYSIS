```python
#Electric vehicle popupation analysis
#intership assignment
#name: Harini J
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
#loading the data#
#The dataset contains information about electric vehicles registered in Washington State, including make, model, year, type, range, and eligibility for clean energy incentives.
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
df.head()
# Data Understanding
# 1.Number of rows & columns: (output of df.shape)
# 2.Columns: (output of df.columns)
#3. Info: Shows data types and missing values
# 4.Describe: Gives statistics of numeric columns
df.shape #number of rows and columns
df.columns #column names
df.info() #data types and missing values
df.describe() # statistical summary
# Data Cleaning
#Missing values were removed to improve data quality.
df.isnull().sum() #checking missing values
df=df.dropna() # remove missing values
# Selected Columns
#Focused on columns important for analysis:
#Model Year, Make, Model, Electric Vehicle Type, Electric Range, Base MSRP, County, City
important_columns = ['Model Year','Make','Model','Electric Vehicle Type','Electric Range','Base MSRP','County','City']
df = df[important_columns]
df.head()
# Data Analysis
# 1. Count of EVs by type
#2. Top 10 EV Brands
#3. Top 10 EV Models
#4. EV Growth by Year
#5. Electric Range Analysis
df['Electric Vehicle Type'].value_counts()
top_makes = df['Make'].value_counts().head(10)
top_makes
df['Model'].value_counts().head(10)
df['Model Year'].value_counts().sort_index()
df['Electric Range'].describe()
         ### Visualizations
#- Bar chart shows top 10 EV brands
#- Pie chart shows distribution of EV types
#- Histogram shows EV model year distribution

df['Make'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 EV Brands")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.show()
df['Electric Vehicle Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("EV Type Distribution")
plt.ylabel("")
plt.show()
plt.hist(df['Model Year'])
plt.title("EV Distribution by Model Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
### Key Insights
#- Tesla is the most popular EV brand
#- Battery Electric Vehicles (BEVs) are more common than Plug-in Hybrid Vehicles (PHEVs)
#- EV adoption increased after 2018
#- Vehicles with higher electric range are mostly newer models
## Conclusion
#The analysis shows rapid growth in EV adoption, with Battery Electric Vehicles dominating the market. Popular brands like Tesla and Hyundai lead in EV registrations. Electric range and pricing show trends in consumer preference.
```
